# SQLite


## Local SQLite DB


```python
import sqlite3
connection = sqlite3.connect("health.db")
cursor = connection.cursor()
```

```python
cursor.execute("""
    CREATE TABLE HEALTHTABLE2(
        Code TEXT primary key,
        Description text,
        Type text,
        Package/Line_Level
        Gross charge float,
        Discounted cash price float,
        De-identified min contracted rate float,
        De-identified max contracted rate float

    )
""")
```
